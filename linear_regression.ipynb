```python
#read the necessary libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

# Using Numpy only 

```python
#create the dataset
x = np.random.rand(500,3)
y = (2*x[:,0]) + (4*x[:,1]) + (1*x[:,2]) + (3*(np.ones(500)))
```

```python
#x vector 
print(x)
```

```python
print(x[:,0].shape)
```

```python
print(y.shape)
```

```python
#visulising the relationship between features and target variable 
plt.figure(figsize=(15,4))
plt.title("Feature vs Target")
plt.xticks([])
plt.yticks([])
for i in range(3):
    plt.subplot(1,3,i+1)
    plt.scatter(x[:,i], y)
    plt.xlabel(f'feature {i}')
plt.show()

```

```python
#adding the bias column to feeatures
ones = np.ones(500)
x = np.column_stack((ones,x))
print(x)
```

```python
#initialize the weights 
weights = np.zeros(4)
print(weights)
```

```python
#gradient descent 
iterations = 1000
lr = 0.05
Loss = []
for epoch in range(iterations):
    y_pred = np.dot(x,weights)
    error = y_pred - y 
    dw = (1/500)*(np.dot(x.T, error))
    weights = weights - (lr*dw)

    loss = (1/500)*(np.sum(error**2))
    Loss.append(loss)
```

```python
print(weights)
```

```python
#loss over iterations
plt.plot(Loss)
plt.title("Loss Function over iterations")
plt.xlabel("Iterations")
plt.ylabel("Loss")
```

# Using Sklearn dataset

```python
#sklearn dataset
x,y = datasets.make_regression(500,n_features=3)
```

```python
print(x)
```

```python
plt.figure(figsize=(15,5))
plt.title("Features vs Target variable")
plt.yticks([])  
plt.xticks([]) 
for i in range(3): 
    plt.subplot(1,3,i+1)
    plt.scatter(x[:,i], y)
    plt.xlabel(f"feature {i}")
plt.tight_layout()
plt.show()
```

```python
#adding bias column 
ones=np.ones(500)
x = np.column_stack((ones,x))
weights = np.zeros(4)
```

```python
print(x.shape)
```

```python
#gradient descent loop
iterations = 1000
n = x.shape[0]
Loss = []
for epoch in range(iterations):
    y_pred = np.dot(x,weights)
    error = y_pred - y
    dw = (1/n)*(np.dot(x.T, error))
    weights = weights - (0.05*dw)

    loss = (1/(2*n))*(np.sum(error**2))
    Loss.append(loss)
```

```python
print(weights)
```

```python
#plotting loss function
plt.plot(Loss)
plt.title("Loss Function over iterations")
plt.xlabel("Iterations")
plt.ylabel("Loss")
```
