```python
ones = np.ones(500)
x = np.column_stack((ones,x))
```

```python
x
```

```python
weights = np.zeros(4)
weights
```

```python
(np.dot(x,weights)).shape
```

```python
iterations = 1000
lr = 0.05
Loss = []
for epoch in range(iterations):
    y_pred = np.dot(x,weights)
    error = y_pred - y 
    dw = (1/500)*(np.dot(x.T, error))
    weights = weights - (lr*dw)

    loss = (1/500)*(np.sum(error**2))
    Loss.append(loss)
```

```python
plt.plot(Loss)
```

```python
ones=np.ones(500)
x = np.column_stack((ones,x))
weights = np.zeros(4)
```

```python
x.shape
```

```python
iterations = 1000
n = x.shape[0]
Loss = []
for epoch in range(iterations):
    y_pred = np.dot(x,weights)
    error = y_pred - y
    dw = (1/n)*(np.dot(x.T, error))
    weights = weights - (0.05*dw)

    loss = (1/(2*n))*(np.sum(error**2))
    Loss.append(loss)
```
